## Data Visualization

- Visual representation of data to discover trends and patterns
- Helps in story telling and decision-makinh

### Types
- Categorical: Bar Chart, Pie Chart, etc
- Numerical: Box Plot, Histogram, etc
- Relationship: Scatter plot, Line plot, Heatmap, Pair plot etc

### Libraries

- Matplotlib
- Seaborn

```python
import pandas as pd

import matplotlib.pyplot as plt
```

- Figure: The entire canvas that holds all subplots
- Axes: A single plot

```python
import matplotlib.pyplot as plt
```

```python
# Create a figure with a single axes

x = [1, 2,3]

y =[4, 5, 6]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Basic Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

plt.show()  #to dispaly your plot
```

```python
#create a figure with multiple subplots

fig, axes = plt.subplots(1, 2) #two plots side by side
axes[0].plot(x, y)
axes[0].set_title("Left Plot")
axes[1].plot(x, y)
axes[1].set_title("Right Plot")

plt.show()
```

```python
fig, axes= plt.subplots(2, 2)

axes[0,0].plot(x,y)
axes[0,1].plot(x,y)
axes[1,0].plot(x,y)
axes[1,1].plot(x,y)

plt.show()
```

## Exploratory Data Analysis (EDA)

- Univariate Analysis: Examining one variable at a time

```python
df= pd.read_csv(r"C:\Users\bunmi\Downloads\netflix_titles (2).csv")

df
```

```python
df.head()
```

```python
df.info()
```

What are the numbers of netflix titles released per year?

```python
title_per_year = df.groupby("release_year")["title"].count().reset_index()

title_per_year
```

```python
#line plot
plt.plot(title_per_year["release_year"], title_per_year["title"], color="red", marker=".")
plt.title("Number of Netflix Titles Released Per Year")
plt.xlabel("Release Year")
plt.ylabel("Numbers of Titles")
plt.show()
```

```python
df["type"].value_counts()
```

```python
type_counts = df["type"].value_counts()
```

```python
type_counts.index
```

```python
type_counts
```

```python
plt.pie(type_counts, labels=type_counts.index, autopct='%1.2f%%', colors=["navy", "lightblue"])
plt.title("Distribution of Movies and TV Shows")

plt.show()
```

```python
df.info()

df["rating"].value_counts()
```

```python
df["rating"] = df["rating"].replace({"74 min":"Unknown", "84 min":"Unknown", "66 min":"Unknown"})
```

```python
rating_counts = df["rating"].value_counts()
```

```python
rating_counts.values
```

```python
rating_counts.index
```

```python
plt.bar(rating_counts.index, rating_counts.values)
plt.xticks(rotation=90)
plt.title("Number of titles per ratings")
plt.xlabel("Ratings")
plt.ylabel("Number of Titles")

plt.show()
```

```python
#histogram is used to show the distribution of a numerical variable

plt.hist(df["release_year"], bins=20)

plt.show()
```

Bivariate analysis

- Examining the relationship between two variables

```python
df["type"].value_counts()
```

```python
#movies and tv shows released per year

release_year_by_type = df.groupby(["release_year", "type"]).size().unstack(fill_value=0).reset_index()

release_year_by_type
```

```python
plt.plot(release_year_by_type["release_year"], release_year_by_type["Movie"], label="Movie")
plt.plot(release_year_by_type["release_year"], release_year_by_type["TV Show"], label="TV Shows")
plt.legend()

plt.show()
```
